```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

MAX_PAGE = 50

def scrape_tribun(max_page):
    data = []
    for page in range(1, max_page + 1):
        print(f"[Tribun] Halaman {page}")
        url = f"https://www.tribunnews.com/index-news?page={page}"
        resp = requests.get(url)
        soup = BeautifulSoup(resp.content, "html.parser")
        articles = soup.select("li.ptb15")
        for a in articles:
            tag = a.find("a")
            if tag:
                data.append({
                    "Sumber": "Tribunnews",
                    "Judul Berita": tag.get_text(strip=True),
                    "Link Judul Berita": tag["href"]
                })
        time.sleep(1)
    return data


def scrape_detik(max_page):
    data = []
    print("[Detik] Memulai scraping...")
    headers = {'User-Agent': 'Mozilla/5.0'}
    for page in range(1, max_page + 1):
        url = f"https://www.detik.com/terpopuler/?_page={page}"
        resp = requests.get(url, headers=headers)
        soup = BeautifulSoup(resp.content, "html.parser")
        articles = soup.select("article a")
        for tag in articles:
            title = tag.get_text(strip=True)
            link = tag['href']
            if title and link and "https://" in link:
                data.append({
                    "Sumber": "Detik",
                    "Judul Berita": title,
                    "Link Judul Berita": link
                })
        time.sleep(1)
    return data



# Eksekusi scraping dengan MAX_PAGE
tribun = scrape_tribun(MAX_PAGE)
detik = scrape_detik(MAX_PAGE)

# Gabungkan data
all_data = tribun + detik
df = pd.DataFrame(all_data)
df.insert(0, "ID", range(1, len(df) + 1))

# Simpan ke CSV
df.to_csv("1000_per_sumber.csv", index=False)

# Tampilkan sampel acak
print(df.sample(10))
```
